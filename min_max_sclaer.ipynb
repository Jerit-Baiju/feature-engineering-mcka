```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = {
    "Brand": [
        "Dell", "HP", "Lenovo", "Apple", "Asus",
        "Acer", "MSI", "Samsung", "Microsoft", "Razer",
        "LG", "Huawei", "Toshiba", "Sony", "Fujitsu",
        "Panasonic", "Gateway", "Alienware", "Chuwi", "Xiaomi",
        "Google", "Vaio", "Sharp", "Dynabook", "iBall"
    ],
    "RAM_GB": [
        8, 16, 8, 32, 16,
        4, 32, 16, 8, 64,
        16, 8, 4, 32, 16,
        8, 16, 32, 4, 8,
        16, 8, 32, 16, 4
    ],
    "Storage_GB": [
        256, 512, 1024, 512, 256,
        128, 2048, 512, 256, 1024,
        512, 256, 128, 2048, 512,
        256, 512, 1024, 128, 256,
        512, 256, 2048, 512, 128
    ],
    "Processor": [
        "i5", "i7", "i5", "M1", "i7",
        "i3", "i9", "i5", "i7", "i9",
        "i5", "i3", "i3", "M2", "i7",
        "i5", "i7", "i9", "i3", "i5",
        "M1", "i5", "M2", "i7", "i3"
    ]
}
```

```python
df = pd.DataFrame(data)
print('Original DataFrame:')
print(df)

# Apply MinMaxScaler to RAM and Storage columns
scaler = MinMaxScaler()
df[['RAM_GB_scaled', 'Storage_GB_scaled']] = scaler.fit_transform(df[['RAM_GB', 'Storage_GB']])

print('\nDataFrame after MinMax Scaling:')
print(df[['Brand', 'RAM_GB', 'RAM_GB_scaled', 'Storage_GB', 'Storage_GB_scaled']])
```
