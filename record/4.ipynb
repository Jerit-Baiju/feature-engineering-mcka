```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create a DataFrame
data = {
    "Name": ["Liam", "Sophia", "Noah", "Olivia", "Ethan", "Ava", "Mason", "Isabella"],
    "Height": [172, 158, 180, 165, 175, 160, 178, 162],
    "Weight": [70, 55, 80, 60, 75, 50, 77, 58],
}

df = pd.DataFrame(data)

# Step 2: Plot histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df["Height"], bins=5, color="skyblue", edgecolor="black")
plt.title("Height Distribution")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(df["Weight"], bins=5, color="lightgreen", edgecolor="black")
plt.title("Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# Step 3: Plot boxplots
plt.figure(figsize=(8, 5))
plt.boxplot(
    [df["Height"], df["Weight"]],
    labels=["Height", "Weight"],
    patch_artist=True,
    boxprops=dict(facecolor="skyblue", color="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    medianprops=dict(color="red"),
)
plt.title("Boxplot of Height and Weight")
plt.ylabel("Value")
plt.show()
```
