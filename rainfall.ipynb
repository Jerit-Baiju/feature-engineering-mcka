# Rainfall Dataset Analysis

This notebook analyzes the rainfall dataset to understand patterns and identify prediction opportunities.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the rainfall dataset
df = pd.read_csv('/Users/jeritbaiju/workspaces/MCKA/feature-engineering-mcka/datasets/rainfall.csv')

# Choose subdivision (example: Kerala)
subdivision = "KERALA"
data = df[df["SUBDIVISION"] == subdivision][["YEAR", "ANNUAL"]]
```

```python
X = data["YEAR"].values.reshape(-1, 1)
y = data["ANNUAL"].values

# Train linear regression
model = LinearRegression()
model.fit(X, y)

# Predict for 2016–2025
future_years = np.arange(2016, 2026).reshape(-1, 1)
future_preds = model.predict(future_years)

# Plot
plt.scatter(X, y, color="blue", label="Actual")
plt.plot(X, model.predict(X), color="red", linewidth=2, label="Trend line")
plt.scatter(future_years, future_preds, color="green", label="Predicted (2016–2025)")

plt.title(f"Rainfall Prediction using Linear Regression ({subdivision})")
plt.xlabel("Year")
plt.ylabel("Annual Rainfall (mm)")
plt.legend()
plt.show()

# Show predictions
for year, pred in zip(future_years.flatten(), future_preds):
    print(f"{year}: {pred:.2f} mm")
```
