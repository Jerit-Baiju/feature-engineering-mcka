```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        "Height": [150, 160, 170, 180, 190],
        "Weight": [50, 60, 70, 80, 90],
    }, columns=["Height", "Weight"]
)
```

```python
# 1. Normalization (L2 norm)
from sklearn.preprocessing import Normalizer, MinMaxScaler

norm = Normalizer()
df_norm = pd.DataFrame(norm.fit_transform(df), columns=df.columns)

# 2. Min-Max scaling [0,1]
scaler = MinMaxScaler()
df_minmax = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print("Normalized:\n", df_norm)
print("Min-Max:\n", df_minmax)
df_norm
```

```python
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({"Income": [1000, 2000, 5000, 10000, 50000]})

# Log transform
df["Log_Income"] = np.log(df["Income"])

# Plot
plt.hist(df["Income"], bins=5)
plt.title("Before Log Transformation")
plt.show()

plt.hist(df["Log_Income"], bins=5)
plt.title("After Log Transformation")
plt.show()
```

```python
df = pd.DataFrame({"Age": [5, 12, 17, 24, 37, 45, 59, 80]})

bins = [0, 12, 19, 59, 120]
labels = ["Child", "Teen", "Adult", "Senior"]

df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels)
print(df)
```

```python
df = pd.DataFrame({"ID": [1, 2, 3, 4, 5], "Color": ["Red", "Blue", "Green", "Red", "Blue"]})

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["Color"])
print("Before:", df.shape[1], "features")
print("After:", df_encoded.shape[1], "features")
print(df_encoded)
```
