```python
import os
import pandas as pd

# Check if you're running on Kaggle
IS_KAGGLE = os.path.exists("/kaggle/input")

if IS_KAGGLE:
    dataset_path = "/kaggle/input/cafe-sales-dirty-data-for-cleaning-training/dirty_cafe_sales.csv"
else:
    dataset_path = "datasets/dirty_cafe_sales.csv"

# Load the dataset
df = pd.read_csv(dataset_path)
display(df.head())
```

```python
# Identify columns with null values
null_counts = df.isnull().sum()
print('Columns with null values:')
print(null_counts[null_counts > 0])
```

```python
# Fill nulls: median for numeric columns, mode for categorical columns
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        median_val = df[col].median()
        df.fillna({col: median_val}, inplace=True)
    else:
        mode_val = df[col].mode().iloc[0] if not df[col].mode().empty else None
        df.fillna({col: mode_val}, inplace=True)

 # Check if nulls remain
print('Null values after filling:')
print(df.isnull().sum())
```
